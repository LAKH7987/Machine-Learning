```python
import pandas as pd
```

```python
df = pd.read_csv("Classification_dataset.csv")
print(df)
```

```python
y_true = df["actual"]
y_pred = df["predicted"]
```

ctual  predicted
1          1   -----> TP
0          0   -----> TN
1          1   -----> TP
1          0   -----> FN
0          0   -----> TN
1          1   -----> TP          TN - 4
0          0   -----> TN          TP - 4
0          1   -----> FP          FP - 1
1          1   -----> TP          FN - 1
0          0   -----> TN
                                 

```python
print(y_true)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
print(confusion_matrix(y_true,y_pred)) #[TN FP]  
                                       #[FN TP]
```

```python
print(accuracy_score(y_true,y_pred))

#Accuracy = (TP+TN)/(FP+FN+TP+TN)
# (4+4) / (1+1+4+4) = 8/10 = 0.8 
# means 80% percent prediction is correct 
```

```python
print(precision_score(y_true,y_pred))

# Precision=TP/(FP+TP)
# 4 / (1+4)  =  0.8,  means total positive prediction ki 80% prediction correct he  (4 - tp and 1 - fp)
```

```python
print(recall_score(y_true,y_pred))

# recall = TP/(TP+FN)
#4/(4+1) - 0.8, means total positive prediction me se 80% percent prediction model ne correct bataye 
```

```python
print(f1_score(y_true,y_pred))

#F1=2TP/(FP+FN+2TP)
# 2*4 / (1+1+2*4) - 8/10 - 0.8, f1 score high because recall and precision are high
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

confuse = confusion_matrix(y_true,y_pred)
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
```
