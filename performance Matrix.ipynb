```python
import pandas as pd
```

```python
df = pd.read_csv("Classification_dataset.csv")
print(df)
```

```python
y_true = df["actual"]
y_pred = df["predicted"]
```

```python
# actual  predicted
# 1          1   -----> TP
# 0          0   -----> TN
# 1          1   -----> TP
# 1          0   -----> FN
# 0          0   -----> TN
# 1          1   -----> TP          TN - 4
# 0          0   -----> TN          TP - 4
# 0          1   -----> FP          FP - 1
# 1          1   -----> TP          FN - 1
# 0          0   -----> TN
                                 
```

```python
print(y_true)
print(y_pred)
```

```python
from sklearn.metrics import *
```

```python
print("confusion matrix is : \n",confusion_matrix(y_true,y_pred)) #[TN FP]  
                                                                  #[FN TP]
```

```python
print("accuracy score is : ",accuracy_score(y_true,y_pred))

#formula - 
#Accuracy = (TP+TN)/(FP+FN+TP+TN)
# (4+4) / (1+1+4+4) = 8/10 = 0.8 

#Interpretation -  Out of all predictions, 80% are correct.
```

```python
print("precision score is : ",precision_score(y_true,y_pred))

#fromula - 
# Precision=TP/(FP+TP)
# 4 / (1+4)  =  0.8,  

# interpretation - Out of all positive predictions, 80% are correct.(4 - tp and 1 - fp)
```

```python
print("recall score is : ",recall_score(y_true,y_pred))

# formula - 
# recall = TP/(TP+FN)
#4/(4+1) - 0.8, 

#interpretation - Out of all the positive predictions made by the model, 80% were correct.”
```

```python
print(f1_score(y_true,y_pred))

#F1=2TP/(FP+FN+2TP)
# 2*4 / (1+1+2*4) - 8/10 - 0.8, 

# interpretation - f1 score high because recall and precision are high
```

```python
fbeta = fbeta_score(y_true, y_pred, beta=1)
print("F-beta score:", fbeta)

# formula - 
#Fβ =(1+β^2)×(Precision×Recall)/(β^2×Precision)+Recall
#(1 × 0.8) + 0.8 = 1.6
#Fβ = 2 × (0.64 / 1.6)
#Fβ = 2 × 0.4
#Fβ = 0.8
	
# interpretation - This means the model performs well in correctly predicting positive cases while also detecting most of the actual positive cases
```

```python
fbeta = fbeta_score(y_true, y_pred, beta=0.5)
print("F-beta score:", fbeta)

# formula - 
#Fβ =(1+β^2)×(Precision×Recall)/(β^2×Precision)+Recall
 #Fβ = ( 
	
# interpretation - The model makes good predictions and focuses more on making correct positive predictions
```

```python
fbeta = fbeta_score(y_true, y_pred, beta=2)
print("F-beta score:", fbeta)

# formula - 
#Fβ =(1+β^2)×(Precision×Recall)/(β^2×Precision)+Recall
 #Fβ  = 1.25 × (0.64 / 1.0)
#  = 1.25 × 0.64
# = 0.8
	
# interpretation - The model predicts positive cases correctly and also detects most of the actual positive cases.
```

```python
confuse = confusion_matrix(y_true,y_pred)
acc = accuracy_score(y_true, y_pred)
ps = precision_score(y_true, y_pred)
rc = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
```
